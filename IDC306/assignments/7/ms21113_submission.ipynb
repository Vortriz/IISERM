### Bubble Sort

```python
a = [random.randint(1, 100) for _ in range(25)]
print(a)
```

```python
def bubblepass(arr) :
    a = arr.copy()

    c = 0
    k = len(a)

    for i in range(k-1) :
        if a[i] > a[i+1] :
            c+=1
            a[i], a[i+1] = a[i+1], a[i]

    return c, a
```

```python
def bubblesort(arr) :
    a = arr.copy()

    c = 1
    while c > 0 :
        c, a = bubblepass(a)

    return a
```

```python
bubblesort(a)
```

### Insert Sort

```python
def insert(a, b) :
	k = len(a)-1
	a.append(b)
	j = k

	while j >= 0 and a[j] > a[j+1] :
		a[j], a[j+1] = a[j+1], a[j]
		j-=1

	return a
```

```python
def insertsort(arr) :
    a = arr.copy()
    k = len(a)

    for i in range(k) :
        a_prime, ai, a_pprime  = a[:i], a[i], a[i+1:]
        a_prime = insert(a_prime, ai)
        a = a_prime + a_pprime

    return a
```

```python
insertsort(a)
```
