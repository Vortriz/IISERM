Properly defining attributes of class entities

```python
class Students:

  def __init__(self,first,last,batch,fees):
    self.fname=first
    self.lname=last
    self.batch=batch
    self.fees=fees

st1=Students("Shanya","Malhotra",'2020-25',10000)

## Direct access to attribute from Class
print(st1.fname)
```

The attributes can be prevented from public access by using '__' infront of the variable.
Defining Instance methods for class objects.

```python
class Students:

  def __init__(self,first,last,batch,fees):
    self.__fname=first
    self.__lname=last
    self.__batch=batch
    self.__fees=fees

  def displayName(self):
    return(self.__fname+' '+self.__lname)

  def displayBatch(self):
    return(self.__batch)

  def displayFees(self):
    return(self.__fees)


st1=Students('Tim','cook','2022-27',10000)
st2=Students('Jensen','Huang','2022-27',10000)

print(st2.displayBatch())
print(st1.displayName()+' '+str(st1.displayFees()))
```

Class method

```python
class Students:

  raiseAmt=1.05
  countStudent=0

  def __init__(self,first,last,batch,fees):
    self.__fname=first
    self.__lname=last
    self.__batch=batch
    self.__fees=fees
    Students.countStudent+=1

  def raiseFeeMethod(self):
    self.__fees=int(self.__fees*self.raiseAmt) ## Class variable but can be changed for an instance

  def displayName(self):
    return(self.__fname+' '+self.__lname)

  def displayBatch(self):
    return(self.__batch)

  def displayFees(self):
    return(self.__fees)

  @classmethod
  def raiseAmount(cls,amt):
    cls.raiseAmt=amt

Students.raiseAmount(1.09)
st1=Students('Tim','cook','2022-27',10000)
st2=Students('Tim','cook','2022-27',10000)

st1.raiseAmt
st2.raiseAmt
```
