```python
def outer(func):
  def inner():
    print('Start')
    func()
    print('End')
  return inner

def greet():
  print('This is getting in inner function')

greet=outer(greet)
print(greet())
```

```python
def outer(func):
  def inner(name):
    print('Start')
    x=func(name)
    print(x)
    print('End')
  return inner

@outer
def greet(name):
  return ('This is getting in inner function '+name)

greet('Test')
```

```python
def decorator(func):
  def wrapper(name):
    x=func(name)
    print(x)
    if name.startswith('IDC'):
      print('This is interdisciplinary course')
    else:
      print('This is subject specific course')
  return wrapper

@decorator
def course(name):
  return ('Welcome to couse '+name)

course('IDC101')
```

```python
class students:
  def __init__(self,name,age,address):
    self.name=name
    self.age=age
    self.address=address

  def getname(self):
    return self.name

class batch(students):
  def __init__(self,name,age,address,batch):
    super().__init__(name,age,address)
    self.batch=batch

x=batch('Test','30','delhi','2021')
print(x.getname(),x.batch)

```

```python
class customError(Exception):
  pass

def check(inp):
  if len(inp) < 8:
    raise customError(' Length less than 8')
  if not inp.startswith('S'):
    raise customError('Does not start with S')

query='jsja'
try:
  check(query)
except customError as error:
  print(error)
```
