@classmethod @staticmethod regular method

```python
class TestClass:

  def rmethod(self):## access to instance /class variables and can modify it
    return('Regular method is accessed')

  @classmethod ## access to class variables and can modify it
  def clmethod(cls):
      return('Class method is accessed')

  @staticmethod
  def stmethod():
    return ('Static method is accessed')

instance = TestClass()
print(instance.rmethod())
print(instance.clmethod())
print(instance.stmethod())

print(TestClass.clmethod())
print(TestClass.stmethod())

print(TestClass.rmethod())## THIS will return error
```

```python
class Students:
  raiseAmt=1.05
  countStudent=0

  def __init__(self,first,last,batch,fees):
    self.fname=first
    self.lname=last
    self.batch=batch
    self.fees=fees
    Students.countStudent+=1

  def raiseFee(self):
    self.fees=int(self.fees*self.raiseAmt)

  def displayName(self):
    return(self.fname+' '+self.lname)

  def displayFees(self):
    return(self.fees)

  @classmethod
  def changeAmt(cls,amt):
    cls.raiseAmt=amt

st1=Students('Tim','cook','2022-27',10000)
st2=Students('Tom','cook','2022-27',10000)

print(st1.raiseAmt)
print(st2.raiseAmt)
```

Various ways of getting and setting attributes. Getter/setter

```python
class ExampleClass:

  def __init__(self,fname,lname):
    self._fname=fname
    self._lname=lname

  def getFname(self):
    return self._fname

  def setFname(self,value):
    self._fname=value

st1=ExampleClass('ram','Shankar')
print(st1.getFname())
st1.setFname('Test')
print(st1.getFname())
```

Getter/setter change attribute

```python
class ExampleClass:

  def __init__(self,fname,lname):
    #self._fname=fname
    self.setFname(fname)
    self._lname=lname

  def getFname(self):
    return self._fname

  def setFname(self,value):
    self._fname=value.upper()

st1=ExampleClass('ram','Shankar')
print(st1.getFname())
st1.setFname('Test')
print(st1.getFname())
```

Getter/setter through @property

```python
class TestClass:

  def __init__(self,fname,lname):
    self.fname=fname
    self._lname=lname

  @property
  def fname(self):
    return self._fname

  @fname.setter
  def fname(self,value):
    self._fname=value.upper()

  @property
  def _lname(self):
    return self.lname

  @_lname.setter
  def _lname(self,value):
    self.lname=value.upper()


st1=TestClass('ram','Shankar')
st1.fname='test'
print(st1.lname,st1.fname)
```
