Decorators: It is a function that modifies another function to extend the behavior of the latter function without explicitly modifying it.

````python
def count(num):
  return num+1
p=count ## here is function reference is passed to p
p(3) ## function is evaluated through reference
````

````python
def sayHello(name):
  return ('Hello '+name)
def sayDay(name):
  return ('Good morning '+name)

def greet(func): ## Note that argument passed to function is FUNCTION REFERENCE
  return func('idc students') ## Function execution is returned

greet(sayDay)
````

Define function (inner) within a function. Inner functions cannot be called from outside. The scope is within its parent function.


````python
def greet():
  def sayHello():
    print( 'Welcome idc ')
  def greetDay():
    print('Good morning idc')
  sayHello()
  greetDay()

greet()
````

````python
def greet(name,msg):
  def sayHello(name):
    return ('Welcome '+name)
  def greetDay(name):
    return ('Good morning '+name)

  if msg == 'welcome':
    return sayHello(name)
  else:
    return greetDay(name)

greet('Tom','welcome')
````

Moving on defining decorators

```
def Name_of_decorator(function):
   def wrapper():
   return wrapper
```

````python
def xmodify(function):
  def wrapper():
    print('before calling in the function')
    function()
    print('after calling the function')
  return wrapper

def sayHello():
  print('Welcome to idc')
sayHello()
m=xmodify(sayHello)
m()
````

Using @ symbol as decorators

````python
@xmodify
def sayHello():
  print('Welcome to idc')

sayHello()
````
