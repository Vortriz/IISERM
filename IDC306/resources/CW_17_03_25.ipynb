Anchors, Backreferencing and lookahead and lookbehind

```python
import re
txt='''
Biology Core Courses
BIO101: Cellular basis of life
BIO111: Biology Lab I
BIO102: Gene expression and development
BIO112: Biology Lab II
BIO201: Genetics and evolution
BIO211: Biology Lab III
BIO202: Behaviour and ecology
BIO212: Biology Lab IV
'''
m=re.findall(r'^([A-Z]{3}\d{3}):\s+(.+)',txt,re.MULTILINE)
print(m)
```

```python
txt='This is test 1\n This is test \n This is test 2\n'
print(txt)
m=re.findall(r'.+\d$',txt,re.MULTILINE)
print(m)
```

```python
# \b
txt='This idc306 course idc306666 course'
m=re.findall(r'\bidc\d{3}\b',txt)
print(m)
```

```python
m=re.search(r'\b(\w+)\b.+\b\1\b','text is repeating text')
print(m)
```

```python
txt='cab:cab'
m=re.search(r'([abc]+)',txt)
print(m,m.groups())
```

```python
txt='idc306 course is repeatng idc306 courte text'
m=re.search(r'(\w+)\s+(\w+).+\b\1\b\s+\b\2\b',txt)
print(m)
```

```python
txt='<html> text is kept for htmling </html>'
m=re.search(r'<(?P<tag>\w+)>(?P<mid>.+)</(?P=tag)>',txt)
print(m,m.groups(),m.groupdict())
```

```python
m=re.search(r'(?P<date>\d\d)-(?P<month>\d\d)-(?P<year>\d-{4})','10-10-2025')
print(m.groupdict())
```
