### 1. Implement the basic randomized algorithms and try to understand their differences.

```julia
function ranfind0_1(a)
    k = length(a)
    t = 0

    while t == 0
        i = rand(1:k)

        if a[i] == 0
            t = i
        end
    end

    return t
end
```

### 2. Implement the randomized quick sort algorithm and compute its average running time by running it many times for the *same* input.

```julia
function cmp(a, b)
    global count_cmp += 1
    return a <= b
end
```

```julia
function partition(a, p)
	k = length(a)
	λ, ρ = Int[], Int[]
	c = a[p]

	for i in 1:k
		if i == p
			continue
		end

		aᵢ = a[i]
		if cmp(aᵢ, c) == true
			push!(λ, aᵢ)
		else
			push!(ρ, aᵢ)
		end
	end

	return λ, ρ
end
```

```julia
function quicksort(a)
	k = length(a)

	 if k <= 1
		 return a
	 end

    p = rand(1:k)
	c = a[p]
	λ, ρ = partition(a, p)
	λ, ρ = quicksort(λ), quicksort(ρ)

	return vcat(λ, c, ρ)
end
```

```julia
function sort_helper(a, f)
    global count_cmp = 0
    time = @elapsed f(a)

    return count_cmp, time
end
```

```julia
list_range = 1000:1000000
list_size = 1000
iters = 1000
```

```julia
a = rand(list_range, list_size)
qs_counts, qs_times = [], []

for i in 1:iters
    count, time = sort_helper(a, quicksort)
    push!(qs_counts, count)
    push!(qs_times, time)
end

qs_avg_count = mean(qs_counts)
qs_avg_time = mean(qs_times)

@show qs_avg_count
@show qs_avg_time
```

### 3. Implement the randomized selection algorithm and do the same as above.

```julia
function ransel(a, m)
    k = length(a)

    if k == 1
        return a
    end

    p = rand(1:k)
    λ, ρ = Int[], Int[]
    for i in 1:k
        aᵢ = a[i]
        if aᵢ <= a[p]
            push!(λ, aᵢ)
        else
            push!(ρ, aᵢ)
        end
    end

    n = length(λ)
    if m <= n
        return ransel(λ, m)
    else
        return ransel(ρ, m - n)
    end
end
```

```julia
a = sample(list_range, list_size, replace=false)
rs_times = []

for i in 1:iters
    time = @elapsed ransel(a, rand(1:list_size))
    push!(rs_times, time)
end

rs_avg_time = mean(rs_times)

@show rs_avg_time
```
